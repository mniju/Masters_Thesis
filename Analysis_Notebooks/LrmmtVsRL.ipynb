```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
lrmmt_path = "D:\\CloudSim2\\MultiLogsRL_Report\\CSVlogs\\OriginalCodeRun\\LrMmt.csv"
lrRL_path = "D:\\CloudSim2\\MultiLogsRL_Report\\CSVlogs\\SARSA_Softmax\\SARSA_softmax_0.4_0.2.csv"
```

```python

df_lr = pd.read_csv(lrmmt_path)
df_RL = pd.read_csv(lrRL_path)
```

```python
def create_ESV(df):
    df['ESV'] = (df["Energy(KWh)"] * df["SLA(%)"])/100
    return df
```

```python
df_lr = create_ESV(df_lr)
df_RL = create_ESV(df_RL)
```

```python
df_RL.head()
```

```python

merged_df = pd.merge(df_lr, df_RL, on=['Days'], how='outer')
```

```python
merged_df.head(20)
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
sns.lineplot(x='Days', y='Energy(KWh)_x', data=merged_df, ax=axs, label='Lr-Mmt')
sns.lineplot(x='Days', y='Energy(KWh)_y', data=merged_df, ax=axs, label='Lr-RL')

# Add a title and labels
axs.set_title('Energy Consumption Lr-Mmt Vs Lr-RL')
axs.set_xlabel('Days')
axs.set_ylabel('Energy(Kwh)')

# Show the legend
axs.legend()
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
sns.lineplot(x='Days', y='Migrations_x', data=merged_df, ax=axs, label='Lr-mmt')
sns.lineplot(x='Days', y='Migrations_y', data=merged_df, ax=axs, label='Lr-RL')

# Add a title and labels
axs.set_title('Total Migrations Lr-Mmt Vs Lr-RL')
axs.set_xlabel('Days')
axs.set_ylabel('No of VM Migrations')

# Show the legend
axs.legend()
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
sns.lineplot(x='Days', y='SLA(%)_x', data=merged_df, ax=axs, label='Lr-mmt')
sns.lineplot(x='Days', y='SLA(%)_y', data=merged_df, ax=axs, label='Lr-RL')

# Add a title and labels
axs.set_title('SLAV% Lr-Mmt Vs Lr-RL')
axs.set_xlabel('Days')
axs.set_ylabel('SLA%')

# Show the legend
axs.legend()
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
sns.lineplot(x='Days', y='ESV_x', data=merged_df, ax=axs, label='Lr-mmt')
sns.lineplot(x='Days', y='ESV_y', data=merged_df, ax=axs, label='Lr-RL')

# Add a title and labels
axs.set_title('ESV Lr-Mmt Vs Lr-RL')
axs.set_xlabel('Days')
axs.set_ylabel('ESV')

# Show the legend
axs.legend()
```
